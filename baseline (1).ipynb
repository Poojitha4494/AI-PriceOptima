```python
import pandas as pd

df = pd.read_csv(r'C:\Users\pooji\OneDrive\Attachments\Desktop\Priceoptima\based_pricing.csv')
```

## Milestone 4 Baseline Rule

```python
# TIME BASED PRICING RULES

df['rule_price'] = df['Price'].astype(float)

# Weekend rule → +10%
df.loc[df['is_weekend'] == 1, 'rule_price'] *= 1.10;
# Month-end rule → +5%
df.loc[df['DAY'] >= 25, 'rule_price'] *= 1.05;
# Festival season rule → +15%
df.loc[df['Season'] == 'Festival', 'rule_price'] *= 1.15;

# final output
df[['Date', 'Price', 'rule_price', 'is_weekend', 'Season', 'DAY', 'Conversion Rate %']]
```

```python
#INVENTORY BASED PRICING RULES

# Start from base price
df['Inventory_Adjusted_Price'] = df['Price'].astype(float)
# 1. Low Stock (<20 → +15%)
df.loc[df['Stock Level'] < 20, 'Inventory_Adjusted_Price'] *= 1.15
```

```python
# 2. High Stock (>100 → -10%)
df.loc[df['Stock Level'] > 100, 'Inventory_Adjusted_Price'] *= 0.90
```

```python
# 3. Overstock (>200 → -25%)
df.loc[df['Stock Level'] > 200, 'Inventory_Adjusted_Price'] *= 0.75
df['Inventory_Adjusted_Price'] = df['Price'].astype(float)
cols_to_show = ['Date','Price','Stock Level','Inventory_Adjusted_Price']

print(df[cols_to_show].sample(10).sort_values(by='Date'))
```

```python
#REVENUE CALCULATIONS

# Static Revenue = original price × units sold
df['Static_Revenue'] = df['Price'] * df['Units Sold']

# Rule-Based Revenue = rule_price × units sold
df['Rule_Based_Revenue'] = df['rule_price'] * df['Units Sold']
```

```python
total_Static_Revenue = df['Static_Revenue'].sum()
total_Rule_Based_Revenue = df['Rule_Based_Revenue'].sum()

print(f"total_Static_Revenue : {total_Static_Revenue:,.2f}")
print(f"total_Rule_Based_Revenue :{total_Rule_Based_Revenue:,.2f}")
```

```python
# Revenue lift (absolute)
revenue_lift = total_Rule_Based_Revenue - total_Static_Revenue
print(f"Revenue Lift : {revenue_lift:,.2f}")

# Revenue lift percentage
revenue_lift_percent = (revenue_lift / total_Static_Revenue) * 100
print(f"Revenue Lift  : {revenue_lift_percent:.2f}%")
```
