```python
print("\nInvalid values check:")
print("Units Sold <=0:",(custom_sales["Units Sold"] <= 0).sum())  
print("Price <=0:",(custom_sales["Price"] <= 0).sum())
print("Revenue <=0:",(custom_sales["Revenue"] <= 0).sum())
print("Cost Price  <=0:",(custom_sales["Cost Price"] <= 0).sum())
print("Stock Level <=0:",(custom_sales["Stock Level"] <= 0).sum())
```

```python
#KPI 1 : Revenue Lift
multiplier = 0.88
custom_sales["Dynamic Price"] = custom_sales["Price"] * multiplier
custom_sales["Dynamic Revenue"] = custom_sales["Dynamic Price"] * custom_sales["Units Sold"]
custom_sales["Expected Revenue"] = custom_sales["Cost Price"] * custom_sales["Units Sold"]
revenue_lift = ((custom_sales["Dynamic Revenue"].sum() - 
                 custom_sales["Expected Revenue"].sum()) /
                custom_sales["Expected Revenue"].sum()) * 100
print(f"Revenue Lift: {revenue_lift:.2f}%")
```

```python
#KPI 2:Profit Margin
revenue_multiplier = 0.85
custom_sales["Adjusted Revenue"] = custom_sales["Revenue"] * revenue_multiplier
custom_sales["Profit"] = custom_sales["Adjusted Revenue"] - (custom_sales["Cost Price"] * custom_sales["Units Sold"])
profit_margin = (custom_sales["Profit"].sum() / custom_sales["Adjusted Revenue"].sum()) * 100
print(f"Profit Margin: {profit_margin:.2f}%")
```

```python
# KPI 3: Conversion Rate
conversion_rate = (custom_sales["Units Sold"].sum() / custom_sales["Stock Level"].sum()) * 100
print(f"Conversion Rate: {conversion_rate:.2f}%")
```

```python
# KPI 4: Inventory Turnover
inventory_turnover = custom_sales["Units Sold"].sum() / custom_sales["Stock Level"].sum()
print(f"Inventory Turnover: {inventory_turnover:.2f}")
```
